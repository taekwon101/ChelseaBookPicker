```python
import pandas as pd
import random
import math

fileLocation = r"ChelseaBooks_-_Sheet1.csv"
columnLabels = ["Title", "Author", "Pages", "Read", "Primary/Secondary"]

df = pd.read_csv(fileLocation, names = columnLabels)
df.sort_values(by=["Pages"])

pageSum = 0
pageGoal = 1600

df_day = pd.DataFrame()
total = 0
used = []

while pageSum < pageGoal:
    num = math.floor(120 * random.random())
    if (df["Read"][num] == "No") and (num not in used) and ((pageSum + int(df["Pages"][num]) < 1800)):
        total += 1
        used.append(num)
        pageSum += int(df["Pages"][num])
        df_day[total] = df.iloc[num]

print('number of books: ',total)
print('total pages: ', pageSum)
print('pages divided by 5: ', pageSum/5)

df_day
```
